```python
from oneargopy.OneArgo import Argo
from argopy import DataFetcher as ArgoDataFetcher
import pandas as pd
import matplotlib.pyplot as plt
```

```python
argo = Argo('argo_config.json')
```

```python
profile = argo.select_profiles(floats=5906583)
df = argo.load_float_data(profile, variables=['DOXY','TEMP','PSAL','CHLA', 'BBP700','NITRATE','PH_IN_SITU_TOTAL'])
```

```python
#convert to datetime
df['DATE'] = pd.to_datetime(df['DATE'],format = 'mixed')
#subset data with good QC only
qc_flags = [1,2,8]
df_clean = df[df['CHLA_ADJUSTED_QC'].isin(qc_flags)] # & df['BBP700_ADJUSTED_QC'].isin(qc_flags)]
```

```python
df_clean
```

```python
df_cycle = df_clean[df_clean['CYCLE_NUMBER'] == 4]
```

```python
df_cycle
```

```python
# Create figure and axis
fig = plt.figure(figsize=(10,6), layout='constrained')
ax = fig.add_subplot(141)
ax.plot(df_cycle['CHLA_ADJUSTED'], df_cycle['PRES_ADJUSTED'], color='green', lw=2)


ax.set_xlim(0, 2)
ax.set_ylim(0, 300)
ax.invert_yaxis()
ax.set_xlabel(r'Chl-a (mg m$^{-3}$)', fontsize=12)
ax.set_ylabel('Pressure (dbar)', fontsize=12)
ax.set_title(' Chl-a Profile Cycle 4', fontsize=14)

ax = fig.add_subplot(142)
ax.plot(df_cycle['BBP700_ADJUSTED'], df_cycle['PRES_ADJUSTED'], color='blue', lw=2)

#ax.set_xlim(0, 2)
ax.set_ylim(0, 300)
ax.invert_yaxis()
ax.set_xlabel(r'Bbp (m$^{-1}$)', fontsize=12)
ax.set_ylabel('Pressure (dbar)', fontsize=12)
ax.set_title(' Backscatter Profile Cycle 4', fontsize=14)

ax = fig.add_subplot(143)
ax.plot(df_cycle['TEMP_ADJUSTED'], df_cycle['PRES_ADJUSTED'], color='red', lw=2)

#ax.set_xlim(0, 2)
ax.set_ylim(0, 300)
ax.invert_yaxis()
ax.set_xlabel(r'Bbp (m$^{-1}$)', fontsize=12)
ax.set_ylabel('Pressure (dbar)', fontsize=12)
ax.set_title(' Temperature Profile Cycle 4', fontsize=14)

ax = fig.add_subplot(144)
ax.plot(df_cycle['PSAL_ADJUSTED'], df_cycle['PRES_ADJUSTED'], color='orange', lw=2)

#ax.set_xlim(0, 2)
ax.set_ylim(0, 300)
ax.invert_yaxis()
ax.set_xlabel(r'Bbp (m$^{-1}$)', fontsize=12)
ax.set_ylabel('Pressure (dbar)', fontsize=12)
ax.set_title(' Salinity Profile Cycle 4', fontsize=14)
```
