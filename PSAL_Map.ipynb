```python
import argopy
from oneargopy.OneArgo import Argo
from argopy import ArgoFloat
```

```python
argo = Argo('argo_config.json')
```

```python
profiles_uswc = argo.select_profiles(lon_lim=[155,170], lat_lim=[-63,-61],type='bgc')
```

```python
psal_data = argo.load_float_data(profiles_uswc, variables=['PSAL'])
```

```python
psal_data
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.feature as cfeature
import cmocean

fig = plt.figure(figsize=(10, 12))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.OCEAN)

sc = ax.scatter(psal_data['LONGITUDE'], psal_data['LATITUDE'], c=psal_data['PSAL_ADJUSTED'], cmap=cmocean.cm.haline, vmin=34.66, vmax=34.75, s=100)



ax.set_extent([153, 172, -60, -64], crs=ccrs.PlateCarree()) # just data region

lat = [-63.164055, -62.79286]
lon = [163.932868, 169.976922]
ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')
ax.set_title('Practical Salinity in AAR Region 2019-2026')
cbar = plt.colorbar(sc, ax=ax, orientation='horizontal', pad=0.03)
cbar.set_label('Practical Salinity (PSU)', labelpad=15)

plt.show()
```
