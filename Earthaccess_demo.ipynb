## Search for data

```python
results = earthaccess.search_data(
    short_name='ATL06',
    bounding_box=(-10, 20, 10, 50),
    temporal=("1999-02", "2019-03"),
    count=10
)
```

## Download files

```python
files = earthaccess.download(results, "./local_folder")
```

## Can also open data in-memory

```python
earthaccess.open(results)
```

## Open a sample file using xarray so that we can examine a variable

```python
ds = xr.open_dataset(files[1], group='/gt1l/land_ice_segments')
ds
```

## Use hvplot to plot the variable and start some data exploration

```python
ds['h_li'].hvplot(kind='scatter', x='latitude', s=2)
```
